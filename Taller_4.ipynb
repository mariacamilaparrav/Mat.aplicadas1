```python
import matplotlib.pyplot as plt
import numpy as np

# Definir la función de Fibonacci
def fibonacci(n):
    series = [0, 1]
    for _ in range(n - 2):
        series.append(series[-1] + series[-2])
    return series[:n]

# Solicitar al usuario el número de términos
n = int(input("Ingrese el número de términos de Fibonacci: "))

# Generar la serie
fib_series = fibonacci(n)

# Graficar la serie
plt.plot(range(n), fib_series, marker='o', linestyle='-', color='b', label='Fibonacci')
plt.xlabel('Índice')
plt.ylabel('Valor')
plt.title(f'Serie de Fibonacci con {n} términos')
plt.legend()
plt.grid()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 100)

seno = np.sin(x)
coseno = np.cos(x)
tangente = np.tan(x)

tangente[np.abs(tangente) > 10] = np.nan

plt.plot(x, seno, label='Seno', color='red')
plt.plot(x, coseno, label='Coseno', color='blue')
plt.plot(x, tangente, label='Tangente', color='green')

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

plt.xlabel('Radianes (x)')
plt.ylabel('Valor')
plt.title('Funciones trigonométricas')
plt.legend()
plt.grid()
plt.show()
```

**Punto 6**

Pendiente de una recta

```python
import matplotlib.pyplot as plt

def pendiente(x1, y1, x2, y2):
    if x1 == x2:
        print("La pendiente es indefinida (división por cero).")
    return (y2 - y1) / (x2 - x1)

def graficar(x1, y1, x2, y2):
    m = pendiente(x1, y1, x2, y2)
    b = y1 - (m * x1)

    valores_x = [x1, x2]
    valores_y = [(m * x) + b for x in valores_x]

    plt.plot(valores_x, valores_y, linestyle='-', color='b')
    plt.scatter([x1, x2], [y1, y2], color='r', zorder=2)

    plt.axhline(0, color='k')
    plt.axvline(0, color='k')
    plt.grid(5)
    plt.title(f'Recta: y = {m}x + ({b})')
    plt.show()

x1 = float(input("Ingrese x1: "))
y1 = float(input("Ingrese y1: "))
x2 = float(input("Ingrese x2: "))
y2 = float(input("Ingrese y2: "))

graficar(x1, y1, x2, y2)
```

# Punto 7

```python
import numpy as np
import matplotlib.pyplot as plt

def proyectil_trayectoria(v0, angulos, g=9.81):

    plt.figure(figsize=(10, 5))

    for angulo in angulos:
        theta = np.radians(angulo)
        t_vuelo = (2 * v0 * np.sin(theta)) / g
        t = np.linspace(0, t_vuelo, num=100)

        x = v0 * np.cos(theta) * t
        y = v0 * np.sin(theta) * t - 0.5 * g * t**2

        plt.plot(x, y, label=f'{angulo}°')

    plt.xlabel("Distancia (m)")
    plt.ylabel("Altura (m)")
    plt.title("Trayectoria de un proyectil para diferentes ángulos")
    plt.legend()
    plt.grid()
    plt.show()

velocidad_inicial = 20
angulos_lanzamiento = [15, 30, 45, 60, 75]

proyectil_trayectoria(velocidad_inicial, angulos_lanzamiento)
```

```python
import numpy as np
import matplotlib.pyplot as plt
# @title Punto 8 { run: "auto", vertical-output: true }
deslizador = 1 # @param {type:"slider", min:1, max:10, step:1}

def graficar(n_maximo):
  x = np.linspace(-3, 3, 200)
  plt.figure(figsize=(8, 5))

  for n in range(1, n_maximo + 1):
        y = x ** n
        plt.plot(x, y, label=f'n={n}')

  plt.xlabel("x")
  plt.ylabel("f(x) = x^n")
  plt.ylim(-3,3)
  plt.title("Funciones de potencia para diferentes valores de n")
  plt.legend()
  plt.grid()
  plt.show()

graficar(int(deslizador))
```
