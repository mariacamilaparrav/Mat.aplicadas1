#I
For each of the following equations, use a graph to obtain initial guesses for solutions and then apply Newton’s method to locate the solutions within 0.0001.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
f = lambda x: x**5 - 6*x**3 + 2*x - 2
g = lambda x: 5*x**4 - 18*x**2 + 2

def newton(f, g, x0, tol=0.0001, iter=5000):
    for i in range(iter):
        fx0 = f(x0)
        fpx0 = g(x0)
        if fpx0 == 0:
            print(f"Derivada cero en x = {x0}")
            return None
        x1 = x0 - (fx0 / fpx0)
        if abs(x1 - x0) < tol:
            return x1
        x0 = x1
    return None
```

```python
x_vals = [x / 100.0 for x in range(-300, 301)]
y_vals = [f(x) for x in x_vals]

plt.plot(x_vals, y_vals)
plt.axhline(0, color='gray')
plt.title("Gráfica de $f(x) = x^5 − 6x^3 + 2x − 2$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```

```python
conjeini = [-2, -1, 2]
roots = []

for conje in conjeini:
    root = newton(f, g, conje)
    if root is not None:
        if root not in roots:
            roots.append(root)

print("Raíces encontradas:")
for r in roots:
    print(r)
```
